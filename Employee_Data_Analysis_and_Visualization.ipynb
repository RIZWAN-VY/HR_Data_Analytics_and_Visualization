## Employee Data Analysis and Visualization

```python
# LIBRARIES
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Loading the Preprocessed Data for Analysis and Visualization

```python
data = pd.read_csv('C:/Rizwan/projects/HR_Data_Analysis_and_Visualization/Processed_HR_Data.csv')
data.head()
```

### Total Employees

```python
employees = data['Name'].count()
print(f"Total Number of Employees : {employees}")
```

### Department Distribution of Employees  
Bar Chart for Department Distribution

```python
plt.figure(figsize=(20,4))
ax = sns.countplot(x='Department', data=data)
plt.xticks(rotation=20, wrap=True)  
plt.xlabel('Department',fontweight='bold')
plt.ylabel('Count',fontweight='bold')
plt.title('Distribution of Employees by Department',fontweight='bold')

for p in ax.patches:
    count = int(p.get_height())  
    ax.annotate(f'{count}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='baseline', fontsize=12, color='black', xytext=(0, 5), textcoords='offset points')

plt.show()
```

### Rating Distribution
Bar Chart for Rating Analysis  

```python
plt.figure(figsize=(20, 4))
ax = sns.countplot(x='Rating', data=data)
plt.xlabel('Rating',fontweight='bold')
plt.ylabel('Count',fontweight='bold')
plt.title('Distribution of Ratings',fontweight='bold')

for p in ax.patches:
    count = int(p.get_height())  
    ax.annotate(f'{count}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='baseline', fontsize=12, color='black', xytext=(0, 5), textcoords='offset points')

plt.show()
```

Performance Rating  
0 = Not Rated : 61 Employees  
1 = Very Poor : 53 Employees  
2 = Poor      : 116 Employees  
3 = Average   : 400 Employees  
4 = Good      : 163 Employees  
5 = Very Good : 82 Employees  

Majority of Employees performance is Average

Calculating Average Rating of Employees Performance

```python
# Calculate Average of Rating
rating_statistics = data['Rating'].mean()
formatted_rating = "{:.2f}".format(rating_statistics) # setting only two digits after decimal point
print(f"Average Performance Rating of employees: {formatted_rating}/5")
```

### Salary Category
Bar Chart for Salary Distribution

```python
salary_category_counts = data['Salary Category'].value_counts()

plt.figure(figsize=(20, 4))
bars = plt.bar(salary_category_counts.index, salary_category_counts.values, color='lightgreen')
plt.xlabel('Salary Category',fontweight='bold')
plt.ylabel('Number of Employees',fontweight='bold')
plt.title('Distribution of Employees by Salary Category',fontweight='bold')
plt.xticks(rotation=0)  

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, yval + 5, int(yval), ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

```python
# Calculate basic statistics
salary_statistics = data['Annual Salary'].describe()
print(f"Statistics of Salary \n{salary_statistics}")
```

Analysis of Salary column:
* Count :- There are 875 employees with salary data.
* Mean :- The average annual salary is approximately 748,899.
* Standard Deviation :- The salary data has a standard deviation of approximately 411,680, indicating some variability.
* Min :- The minimum annual salary is 281,300.
* 25th Percentile (Q1) :- 25% of employees have a salary less than or equal to 506,250.
* Median (50th Percentile or Q2) :- The median salary is approximately 742,800, which is the middle value when all salaries are sorted.
* 75th Percentile (Q3) :- 75% of employees have a salary less than or equal to 9,679,500.
* Max :- The maximum annual salary is a substantial 10,000,000.

### Tenure in Company - Years of Experience of Employees  
Bar Chart for Experience  

```python
Experience_counts = data['Experience'].value_counts()

plt.figure(figsize=(20, 4))
bars = plt.bar(Experience_counts.index, Experience_counts.values, color='cyan')
plt.xlabel('Years of Experience',fontweight='bold')
plt.ylabel('Number of Employees',fontweight='bold')
plt.title('Employees Experience',fontweight='bold')
plt.xticks(rotation=0)  # Rotate x-axis labels if needed

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, yval + 5, int(yval), ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

### GENDER DIVERSITY ANALYSIS  
Bar Chart and the count of employees by gender  

```python
# Calculate the count of employees by gender
gender_distribution = data['Gender'].value_counts()

# Visualize gender distribution 
plt.figure(figsize=(20, 4))
plt.bar(gender_distribution.index, gender_distribution.values)
plt.title('Gender Distribution in the Company',fontweight = 'bold')
plt.xlabel('Gender',fontweight = 'bold')
plt.ylabel('Count',fontweight = 'bold')
plt.show()

print(gender_distribution)
```

```python
# Analyze Impact on Salary, Tenure, and Performance Ratings:
# Group data by gender and calculate statistics
gender_stats = data.groupby('Gender').agg({
    'Annual Salary': ['mean', 'median', 'std'],
    'Tenure in Company': 'mean',
    'Rating':'mean'})

print(gender_stats)
```

```python
# Gender Pay Gaps :
# Calculate the average salary for each gender
average_salary_by_gender = data.groupby('Gender')['Annual Salary'].mean()
# Calculate the gender pay gap
gender_pay_gap = average_salary_by_gender['Male'] - average_salary_by_gender['Female']
print(f"Gender Pay Gap (Male - Female): {gender_pay_gap:.2f} INR")
```
