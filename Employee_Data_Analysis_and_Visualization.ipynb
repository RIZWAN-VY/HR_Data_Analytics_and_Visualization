## Employee Data Analysis and Visualization

```python
# LIBRARIES
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Loading the Preprocessed Data for Analysis and Visualization

```python
data = pd.read_csv('C:/Rizwan/projects/HR_Data_Analysis_and_Visualization/Processed_HR_Data.csv')
data.head()
```

### Total Employees

```python
employees = data['Name'].count()
print(f"Total Number of Employees : {employees}")
```

### Department Distribution of Employees  
Bar Chart for Department Distribution

```python
plt.figure(figsize=(15,5))
ax = sns.countplot(x='Department', data=data)
plt.xticks(rotation=20, wrap=True)  
plt.xlabel('Department',fontweight='bold')
plt.ylabel('Count',fontweight='bold')
plt.title('Distribution of Employees by Department',fontweight='bold')

for p in ax.patches:
    count = int(p.get_height())  
    ax.annotate(f'{count}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='baseline', fontsize=12, color='black', xytext=(0, 5), textcoords='offset points')

plt.show()
```

### Rating Distribution
Bar Chart for Rating Analysis  

```python
plt.figure(figsize=(15, 5))
ax = sns.countplot(x='Rating', data=data)
plt.xlabel('Rating',fontweight='bold')
plt.ylabel('Count',fontweight='bold')
plt.title('Distribution of Ratings',fontweight='bold')

for p in ax.patches:
    count = int(p.get_height())  
    ax.annotate(f'{count}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='baseline', fontsize=12, color='black', xytext=(0, 5), textcoords='offset points')

plt.show()
```

Performance Rating  
0 = Not Rated : 61 Employees  
1 = Very Poor : 53 Employees  
2 = Poor      : 116 Employees  
3 = Average   : 400 Employees  
4 = Good      : 163 Employees  
5 = Very Good : 82 Employees  

Majority of Employees performance is Average

Calculating Average Rating of Employees Performance

```python
# Calculate Average of Rating
rating_statistics = data['Rating'].mean()
formatted_rating = "{:.2f}".format(rating_statistics) # setting only two digits after decimal point
print(f"Average Performance Rating of employees: {formatted_rating}/5")
```

### Salary Category
Bar Chart for Salary Distribution

```python
salary_category_counts = data['Salary Category'].value_counts()

plt.figure(figsize=(15, 5))
bars = plt.bar(salary_category_counts.index, salary_category_counts.values, color='lightgreen')
plt.xlabel('Salary Category',fontweight='bold')
plt.ylabel('Number of Employees',fontweight='bold')
plt.title('Distribution of Employees by Salary Category',fontweight='bold')
plt.xticks(rotation=0)  

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, yval + 5, int(yval), ha='center', va='bottom')

plt.tight_layout()
plt.show()
```
